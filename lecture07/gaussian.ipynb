```python
# 파이썬 유틸리티 함수 정의
import matplotlib.pyplot as plt
import numpy as np

def plot_image(image, title=''):
    plt.imshow(image)
    plt.axis('off')
    plt.title(title, size=20)

def SaltPepperNoise(im, percent):
    imTmp = im.copy()
    n = int(im.width * im.height * percent / 100)
    x = np.random.randint(0, im.width, n)
    y = np.random.randint(0, im.height, n)
    for (x,y) in zip(x,y):
        px = ( 0 if (np.random.rand() < 0.5) else 255)
        imTmp.putpixel((x,y), px)
    return imTmp
```

```python
# 가우시안 필터링
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

imIn = Image.open("cameraman.jpg").convert('L')
# Gaussian Kernel
imOut = imIn.filter(ImageFilter.GaussianBlur(3))

# Noise & Filtering
imNoise = SaltPepperNoise(imIn, 5)
imFilter = imNoise.filter(ImageFilter.GaussianBlur(3))

plt.figure(figsize=(20,15))
plt.gray()
plt.subplot(141), plot_image(imIn, 'Source')
plt.subplot(142), plot_image(imOut, 'Gaussian')
plt.subplot(143), plot_image(imNoise, 'Noise')
plt.subplot(144), plot_image(imFilter, 'Filtered')
plt.show()
```
