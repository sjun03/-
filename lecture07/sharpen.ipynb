```python
# 파이썬 유틸리티 함수 정의
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.color import rgb2gray

def plot_image(image, title=''):
    plt.imshow(image)
    plt.axis('off')
    plt.title(title, size=20)
```

```python
from numpy import pi
from scipy import signal

def LoG(ksize=12, sigma=3):
    k = ksize
    s = sigma
    n = 2*k+1 # size of the kernel
    kernel = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            kernel[i,j] = -(1-((i-k)**2+(j-k)**2)/(2.*s**2))*np.exp(-((i-k)**2+(j-k)**2)/(2.*s**2))/(pi*s**4)
    kernel = np.round(kernel / np.sqrt((kernel**2).sum()),3)
    return kernel

def DoG(ksize=12, sigma=3):
    k = ksize
    s = sigma
    n = 2*k+1 # size of the kernel
    s1, s2 = s * np.sqrt(2), s / np.sqrt(2)
    kernel = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            kernel[i,j] = np.exp(-((i-k)**2+(j-k)**2)/(2.*s1**2))/(2*pi*s1**2) - np.exp(-((i-k)**2+(j-k)**2)/(2.*s2**2))/(2*pi*s2**2)
    kernel = np.round(kernel / np.sqrt((kernel**2).sum()),3)
    return kernel

def plot_kernel(kernel, ksize, title=''):
    plt.title(title+" Kernel Size: "+ str(ksize), size=20)
    plt.imshow(kernel, cmap='gray')
    # plt.imshow(kernel, cmap='afmhot')
    plt.axis('off')
```

```python
from scipy import signal

sig = 3
ks = 12
inImg = rgb2gray(imread('parrot.png'))
kernel = LoG(ksize = ks, sigma = sig)
outImg = signal.convolve2d(inImg, kernel)

plt.gray()
plt.figure(figsize=(10,10))
plt.subplot(221), plot_kernel(kernel, ks, 'LoG')
plt.subplot(222), plot_image( np.clip(outImg, 0, 1), 'output image with LoG')

kernel = DoG(ksize = ks, sigma = sig)
outImg = signal.convolve2d(inImg, kernel)
plt.subplot(223), plot_kernel(kernel, ks, 'DoG')
plt.subplot(224), plot_image( np.clip(outImg, 0, 1), 'output image with DoG')

plt.show()
```

```python
from scipy import ndimage
from scipy import signal

im = rgb2gray(imread('dip.png'))
im_blurred = ndimage.gaussian_filter(im, 5)
im_umask = np.clip(im - im_blurred, 0, 1) # unsharp mask

plt.gray()
plt.figure(figsize=(20,10))
plt.subplot(231), plot_image(im, 'Original Image')
plt.subplot(232), plot_image(im_blurred, 'Blurred Image')
plt.subplot(233), plot_image(im_umask, 'Unsharp mask')

alpha = [1, 5, 10]
for i in range(3):
  im_sharp = np.clip(im + alpha[i]*im_umask, 0, 1)
  plt.subplot(2,3,i+4)
  plot_image(im_sharp, "Sharpened image, alpha="+ str(alpha[i]))

plt.show()
```
