```python
# 파이썬 유틸리티 함수 정의
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.color import rgb2gray

def plot_image(image, title=''):
    plt.imshow(image)
    plt.axis('off')
    plt.title(title, size=20)
```

```python
import numpy as np
import cv2
from PIL import Image

# cv::COLORMAP_AUTUMN = 0, cv::COLORMAP_BONE = 1, cv::COLORMAP_JET = 2,
# cv::COLORMAP_WINTER = 3, cv::COLORMAP_RAINBOW = 4, cv::COLORMAP_OCEAN = 5
# cv::COLORMAP_SUMMER = 6, cv::COLORMAP_SPRING = 7, cv::COLORMAP_COOL = 8
# cv::COLORMAP_HSV = 9, cv::COLORMAP_PINK = 10, cv::COLORMAP_HOT = 11
# cv::COLORMAP_PARULA = 12, cv::COLORMAP_MAGMA = 13, cv::COLORMAP_INFERNO = 14
# cv::COLORMAP_PLASMA = 15,  cv::COLORMAP_VIRIDIS = 16, cv::COLORMAP_CIVIDIS = 17
# cv::COLORMAP_TWILIGHT = 18, cv::COLORMAP_TWILIGHT_SHIFTED = 19, cv::COLORMAP_TURBO = 20,
# cv::COLORMAP_DEEPGREEN = 21

gray_img = cv2.imread('lena.jpg', cv2.IMREAD_GRAYSCALE)
pseudo_img = cv2.applyColorMap(gray_img, cv2.COLORMAP_JET)
pseudocolored_img = cv2.cvtColor(pseudo_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.gray()
plt.subplot(121), plot_image(gray_img, 'gray')
plt.subplot(122), plot_image(pseudocolored_img, 'pseudocolored')
plt.show()
```

```python
import numpy as np
import cv2
from PIL import Image

img = cv2.imread('chroma.jpg', cv2.IMREAD_COLOR)
img2 = cv2.imread('beach.png', cv2.IMREAD_COLOR)

converted = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
greenScreen = converted.copy()

greenScreen = cv2.inRange(converted, (60-10, 100, 100), (60+10, 255, 255))

inverted = cv2.bitwise_not(greenScreen)
dst = cv2.bitwise_and(img, img, mask = inverted)
dst1 = cv2.bitwise_or(dst, img2, mask = greenScreen)
result = cv2.bitwise_or(dst, dst1)
output = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,10))
plt.gray()
plt.subplot(221), plot_image(img, 'Input')
plt.subplot(222), plot_image(greenScreen, 'Mask')
plt.subplot(223), plot_image(dst, 'Maksed')
plt.subplot(224), plot_image(output, 'Result')
plt.show()
```

```python
import numpy as np
import cv2
from PIL import Image

img = cv2.imread('ball.jpg', cv2.IMREAD_COLOR)
imgshow = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
converted = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
masked = converted.copy()

# green : (29, 86, 6), (64, 255, 255)
# blue : (57, 68, 0), (151, 255, 255)
masked = cv2.inRange(converted, (29, 86, 6), (64, 255, 255))

plt.figure(figsize=(10,10))
plt.gray()
plt.subplot(121), plot_image(imgshow, 'Input')
plt.subplot(122), plot_image(masked, 'Mask')
plt.show()
```
