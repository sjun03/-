```python
# 파이썬 유틸리티 함수 정의
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.color import rgb2gray

def plot_image(image, title=''):
    plt.imshow(image)
    plt.axis('off')
    plt.title(title, size=20)
```

```python
import numpy as np
import cv2
from PIL import Image

RGB_img = Image.open("parrot.png")

white = np.array([255, 255, 255], np.uint8)
CMY_img = white - RGB_img
Cyan, Magenta, Yellow = cv2.split(CMY_img) # 채널 분리

titles = ['RGB_img','CMY_img','Cyan','Magenta','Yellow']

plt.figure(figsize=(20,10))
plt.gray()
plt.subplot(231), plot_image(RGB_img, titles[0])
plt.subplot(232), plot_image(CMY_img, titles[1])
plt.subplot(234), plot_image(Cyan, titles[2])
plt.subplot(235), plot_image(Magenta, titles[3])
plt.subplot(236), plot_image(Yellow, titles[4])
plt.show()
```

```python
import numpy as np
import cv2
from PIL import Image

RGB_img = Image.open("parrot.png")

white = np.array([255, 255, 255], np.uint8)
CMY_img = white - RGB_img
CMY = cv2.split(CMY_img) # 채널 분리

black = cv2.min(CMY[0], cv2.min(CMY[1], CMY[2])) # 원소 간의 최솟값 저장
Cyan, Magenta, Yellow = CMY - black

titles = ['RGB', 'black', 'Cyan','Magenta','Yellow']

plt.figure(figsize=(20,10))
plt.gray()
plt.subplot(231), plot_image(RGB_img, titles[0])
plt.subplot(232), plot_image(black, titles[1])
plt.subplot(234), plot_image(Cyan, titles[2])
plt.subplot(235), plot_image(Magenta, titles[3])
plt.subplot(236), plot_image(Yellow, titles[4])
plt.show()
```

```python
import numpy as np
import cv2
import math

def calc_hsi(bgr):
    # B, G, R = bgr.astype(float)                           # float 형 변환
    B, G, R = float(bgr[0]), float(bgr[1]), float(bgr[2])       # 속도면에 유리
    bgr_sum = (R + G + B)
    # 색상 계산
    tmp1 = ((R - G) + (R - B)) * 0.5
    tmp2 = math.sqrt((R - G) * (R - G) + (R - B) * (G - B))
    angle = math.acos(tmp1 / tmp2) * (180 / np.pi) if tmp2 else 0

    H = angle if B <= G else 360 - angle
    S = 1.0 - 3 * min([R, G, B]) / bgr_sum if bgr_sum else 0
    I = bgr_sum / 3                                                # 명도 계산
    return (H/2, S*255, I)

# BGR 컬러 -> HSI 컬러
def bgr2hsi(image):
    hsv = [[calc_hsi(pixel) for pixel in row] for row in image ]   # 2차원 배열 순회
    return (np.array(hsv)).astype('uint8')

BGR_img = cv2.imread("color_space.jpg", cv2.IMREAD_COLOR) # 컬러 영상 읽기
RGB_img = cv2.cvtColor(BGR_img, cv2.COLOR_BGR2RGB)         # BGR을 RGB로 변환

HSI_img = bgr2hsi(BGR_img)                  # BGR를 HSI로 변환
HSV_img = cv2.cvtColor(BGR_img, cv2.COLOR_BGR2HSV) # OpenCV 함수

Hue, Saturation, Intensity = cv2.split(HSI_img)             # 채널 분리
Hue2, Saturation2, Intensity2 = cv2.split(HSV_img) 					# 채널 분리

titles = ['RGB_img','Hue','Saturation','Intensity']
plt.figure(figsize=(20,10))
plt.gray()
plt.subplot(241), plot_image(RGB_img, titles[0])
plt.subplot(242), plot_image(Hue, titles[1])
plt.subplot(243), plot_image(Saturation, titles[2])
plt.subplot(244), plot_image(Intensity, titles[3])

plt.subplot(245), plot_image(RGB_img, titles[0])
plt.subplot(246), plot_image(Hue2, titles[1])
plt.subplot(247), plot_image(Saturation2, titles[2])
plt.subplot(248), plot_image(Intensity2, titles[3])
plt.show()
```
