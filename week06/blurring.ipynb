```python
# 파이썬 유틸리티 함수 정의
import matplotlib.pyplot as plt
import numpy as np

def plot_image(image, title=''):
    plt.imshow(image)
    plt.axis('off')
    plt.title(title, size=20)
```

```python
# Box Kernel 정의하기
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

kernelSize = 5
boxKernel = np.ones(kernelSize*kernelSize) / (kernelSize*kernelSize)

imIn = Image.open("pattern.jpg").convert('L')
imOut = imIn.filter(ImageFilter.Kernel((kernelSize, kernelSize), boxKernel))

plt.figure(figsize=(10,5))
plt.gray()
plt.subplot(121), plot_image(imIn, 'Source')
plt.subplot(122), plot_image(imOut, 'Filtered')
plt.show()
```

```python
# Box Kernel 정의하기
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

imIn = Image.open("pattern.jpg").convert('L')
imOut = imIn.filter(ImageFilter.Kernel((kernelSize, kernelSize), boxKernel))

plt.figure(figsize=(15,10))
plt.gray()
plt.subplot(1, 5, 1), plot_image(imIn, 'Source')

d = 1
for n in [3, 5, 7, 9]:
  imOut = imIn.filter(ImageFilter.BoxBlur(n))
  plt.subplot(1, 5, d+1)
  plot_image(imOut, 'Box Size:'+str(n))
  d += 1

plt.show()
```
