```python
# 파이썬 유틸리티 함수 정의
import matplotlib.pyplot as plt
import numpy as np

def plot_image(image, title=''):
    plt.imshow(image)
    plt.axis('off')
    plt.title(title, size=20)
```

```python
# PIL 에 정의되어 있는 다양한 필터
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt

imIn = Image.open("cameraman.jpg")
imOut = imIn.filter(ImageFilter.BLUR)
#imOut = imIn.filter(ImageFilter.SMOOTH)
#imOut = imIn.filter(ImageFilter.SHARPEN)
#imOut = imIn.filter(ImageFilter.EMBOSS)

plt.figure(figsize=(10,5))
plt.gray()
plt.subplot(121), plot_image(imIn, 'Source')
plt.subplot(122), plot_image(imOut, 'Filtered')
plt.show()
```

```python
# 커널 정의하기
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

kernelSize = 3
kernel = np.array([1, 1, 1, 1, 1, 1, 1, 1, 1]) / (kernelSize*kernelSize)
#kernel = np.ones(kernelSize*kernelSize) / (kernelSize*kernelSize)

imIn = Image.open("cameraman.jpg")
imOut = imIn.filter(ImageFilter.Kernel((kernelSize, kernelSize), kernel))

plt.figure(figsize=(10,5))
plt.gray()
plt.subplot(121), plot_image(imIn, 'Source')
plt.subplot(122), plot_image(imOut, 'Filtered')
plt.show()
```
